```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import tensorflow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Flatten, Dense
```

### Data Loading & Preprocessing

```python
train_path = 'C:/Users/JAYAKUMARV/Desktop/Self-Learning/Image Analytics/Image Classification/Kaggle/cotton_disease_images/train'
test_path = 'C:/Users/JAYAKUMARV/Desktop/Self-Learning/Image Analytics/Image Classification/Kaggle/cotton_disease_images/val'
```

```python
### Creating ImageDataGenerator for training images augmentation
train_datagen = ImageDataGenerator(rotation_range=90,
                                  width_shift_range=0.2,
                                  height_shift_range=0.2,
                                  zoom_range=0.2,
                                  horizontal_flip=True,
                                  vertical_flip=True,
                                  rescale=1.0/255)

### Creating ImageDataGenerator for testing images augmentation
test_datagen = ImageDataGenerator(rescale=1.0/255)
```

```python
### Generating Augmented Training Images
train_newdata = train_datagen.flow_from_directory(train_path,
                                                  target_size=(224,224),
                                                  batch_size=55,
                                                  shuffle=True,
                                                  class_mode='categorical')
```

```python
### Generating Augmented Testing Images
test_newdata = test_datagen.flow_from_directory(test_path,
                                               target_size=(224,224),
                                               batch_size=55,
                                               shuffle=True,
                                               class_mode='categorical')
```

```python
train_newdata.class_indices
```

```python
test_newdata.class_indices
```

### Model Building (Inception V3)

```python
inception = InceptionV3(input_shape=[224,224,3],
                       weights='imagenet',
                       include_top=False)
```

```python
### Do not train existing weights!
for layer in inception.layers:
    layer.trainable = False
```

```python
x = Flatten()(inception.output)
```

```python
prediction = Dense(4, activation='softmax')(x)
```

```python
### Creating a model (Inception V3 + Customized Bottom Layers)
model = Model(inputs=inception.input, outputs=prediction)
```

```python
### Getting Inception V3 Model Network
model.summary()
```

```python
model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
```

### Training the Inception V3 Model

```python
history = model.fit_generator(train_newdata,
                        validation_data=test_newdata,
                        epochs=10,
                        steps_per_epoch=len(train_newdata),
                        validation_steps=len(test_newdata))
```

```python
# Plot Loss (Training vs Validation)
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()

# Plot Accuracy (Training vs Validation)
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
```

```python
# Saving Model
model.save('Jaya_InceptionV3.h5')
```

```python
# Loading Model
model = load_model('Jaya_InceptionV3.h5')
```

### Predicting on Real Time Image

```python
def image_predict(filename):
    img = image.load_img(filename, target_size=(224,224))
    x = image.img_to_array(img)
    x = x/255
    x = np.expand_dims(x, axis=0)
    img_data = preprocess_input(x)
    pred = np.argmax(model.predict(img_data), axis=1)
    return pred
```
